#Tricky: Create the sentiment model

## Hint: Create a new column called target column that only holds binary data

```python
train_data,test_data = product_final.random_split(.8, seed=0)
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
analyze_these_columns = ['number_of_awesome_words','number_of_great_words','number_of_fantastic_words','number_of_amazing_words','number_of_love_words','number_of_horrible_words','number_of_bad_words','number_of_terrible_words','number_of_awful_words','number_of_wow_words','number_of_hate_words']
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                     target='target_column',
                                                     features=analyze_these_columns,
                                                     validation_set=test_data)
```

```python
selected_words_model['coefficients']['value'].sort()
```

```python
selected_words_model['coefficients']
```

##Compare the accuracy of different sentiment analysis

```python
selected_words_model.evaluate(test_data,metric='roc_curve')
```

```python
selected_words_model.show(view='Evaluation')
```

## Interpreting the difference in performance between models

```python
diaper_champ_reviews = product_final[product_final['name'] == 'Baby Trend Diaper Champ']
```

```python
train_data,test_data = product_final.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                     target='target_column',
                                                     features=['word_count'],
                                                     validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data,metric='roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```

```python
diaper_champ_reviews['predicted_sentiment'] = sentiment_model.predict(diaper_champ_reviews, output_type='probability')
```

```python
diaper_champ_reviews.head()
```

```python
diaper_champ_reviews = diaper_champ_reviews.sort('predicted_sentiment', ascending=False)
```
