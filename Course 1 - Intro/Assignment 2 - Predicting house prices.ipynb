#Load some house sales data

Dataset is from house sales in King County, the region where the city of Seattle, WA is located.

#Exploring the data for housing sales 

The house price is correlated with the number of square feet of living space.

```python
graphlab.canvas.set_target('ipynb')
sales.show(view="Scatter Plot", x="sqft_living", y="price")
```

#Create a simple regression model of sqft_living to price

Split data into training and testing.  
We use seed=0 so that everyone running this notebook gets the same results.  In practice, you may set a random seed (or let GraphLab Create pick a random seed for you).  

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

##Build the regression model using only sqft_living as a feature

```python
sqft_model = graphlab.linear_regression.create(train_data, target='price', features=['sqft_living'])
```

#Evaluate the simple model

```python
print test_data['price'].mean()
```

```python
print sqft_model.evaluate(test_data)
```

RMSE of about \$255,170!

## Note 2
In the module we discussed residual sum of squares (RSS) as an error metric for regression, but GraphLab Create uses root mean squared error (RMSE). These are two common measures of error regression, and RMSE is simply the square root of the RSS:

RMSE can be more intuitive than RSS, since its units are the same as that of the target column in the data, in our case the unit is dollars ($).

```python
train_data,test_data = sales.random_split(.8,seed=0)
```

```python
my_features_advanced_model = graphlab.linear_regression.create(train_data,target='price',features=my_features_advanced,validation_set=None)
```

```python
my_features_original_model = graphlab.linear_regression.create(train_data,target='price',features=my_features_original,validation_set=None)
```

```python
print my_features_advanced_model.evaluate(test_data)
print my_features_original_model.evaluate(test_data)
```
