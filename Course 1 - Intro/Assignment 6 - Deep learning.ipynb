```python
def something(test_set):    
    return knn_model_dog.query(test_set, k=1)['distance']
```

```python
dog_distances.sort('dog-bird',ascending=True)
```

##Define the function

```python
import operator
def is_dog_correct(row):
    return sorted(row.items(), key=operator.itemgetter(1))[0][0] == 'dog-dog'
```

```python
dog_distances.apply(is_dog_correct).sum()
```
